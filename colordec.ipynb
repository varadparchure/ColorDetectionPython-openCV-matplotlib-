```python
VARAD PARCHURE

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("C:/Users/ROG/Desktop/colorpic.jpg")
```

```python

plt.imshow(img)
```

```python
#here colours are in bgr format so convert it into rgb format

RGBimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGBimg)
```

```python
#opencv uses hsv color format insted of rgb or bgr

HSVimg = cv2.cvtColor(RGBimg, cv2.COLOR_RGB2HSV)
plt.imshow(HSVimg)
```

```python
#NOW WE NEED LOWER AND UPPER HSV COLOR RANGE TO DETECT THE COLOR
#hsv is hue->color,saturation->strong ness,value->brightness

#lets mask all color and keep only red color

lower = np.array([155,25,0])
upper = np.array([179,255,255])

mask=cv2.inRange(HSVimg,lower,upper)
```

```python
plt.imshow(mask)
#mask is shown in yello color
```

```python
res = cv2.bitwise_and(RGBimg, RGBimg, mask=mask)
plt.imshow(res)
```

```python
#lets do this for blue color


lower_blue = np.array([110,50,50])
upper_blue = np.array([130,255,255])

mask1=cv2.inRange(HSVimg,lower_blue,upper_blue)
```

```python
plt.imshow(mask1)
```

```python

res1 = cv2.bitwise_and(RGBimg, RGBimg, mask=mask1)
plt.imshow(res1)
```
